```
import os
import numpy as np
from pydub import AudioSegment

sr = 8000 #16000 sampling rate
```

```
speech_key = []
speech = {}

DR = '/content/drive/MyDrive/'

for i in range(1,10):
  avs = DR + "avspeech" + str(i)
  for root, dirs, files in os.walk(avs):
    for filename in files:
        if filename not in speech_key:
          speech[filename] = os.path.join(root,filename)
          speech_key.append(filename)
#got all list from avspeech1-10

len(speech)
#this is training set
```

```
num = 20000
total = len(speech)
spk2_start = total - num

#speech_key[i] has filename
for i in range(num):
  spk1 = speech[speech_key[i]]
  spk2 = speech[speech_key[i+spk2_start]]

  sound1 = AudioSegment.from_file(spk1)
  sound2 = AudioSegment.from_file(spk2)
  frame1 = sound1.frame_count()
  frame2 = sound2.frame_count()
  frame = min(max(frame1, frame2, 3*sr), 4*sr)#min3 max 4 sec
  msec = int(1000*frame/sr)

  delay1 = 0#(np.random.rand()/2).__round__(3)
  delay2 = 0#(np.random.rand()/2).__round__(3)#delay 0~0.5s
  output = AudioSegment.silent(duration = msec, frame_rate=sr)
  sound1 = output.overlay(sound1, position = delay1*1000).set_channels(1)
  sound2 = output.overlay(sound2, position = delay2*1000).set_channels(1)

  dB1 = sound1.dBFS
  dB2 = sound2.dBFS

  maxdB = max(dB1,dB2,-20)
  dB1 = maxdB-dB1
  dB2 = maxdB-dB2

  sound1 = sound1 + dB1
  sound2 = sound2 + dB2
  output = sound1.overlay(sound2)

  spk1_n = base + mid[0] + "/" + spk[0] + "/" + speech_key[i]
  spk2_n = base + mid[0] + "/" + spk[1] + "/" + speech_key[i+spk2_start]
  output_name = base + mid[0] + "/" + spk[2] + "/" + speech_key[i][:-4] + "_" + speech_key[i+spk2_start]

#  print(output_name)
#  print(spk1_n)
#  print(spk2_n)
  sound1.export(spk1_n, format='wav')
  sound2.export(spk2_n, format='wav')
  output.export(output_name, format='wav')
  
  if(i%200 == 199):
    print(i+1, " done")

print("tr done")
```
